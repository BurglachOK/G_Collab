Підкидають три гральні кубики. Що ймовірніше: отримати в сумі очок, що випали, 11 або 12? Розв’яжи програмно, згенерувавши всі можливі комбінації.

```python
import random
def combination():
    eleven = 0
    twelve = 0
    for i in range(1,7):
        for j in range(1,7):
            for k in range(1,7):
                if i + j + k == 11:
                    eleven+=1
                elif i + j + k == 12:
                    twelve+=1
    print(eleven)
    print(twelve)
combination()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np
def randy_random(amount):
    quantity = []
    price = 0
    for number in range(amount):
        for i in range(3):
            price += random.choice((-1, 2))
        quantity.append(price)
        price = 0
    return quantity

def avg(quantity):
    midle = sum(quantity) / len(quantity)
    return midle


# for i in range(1,5):
#     data = randy_random(10**i)
#     print(avg(data))
#     plt.hist(data, bins = 4, alpha = 0.7, color ='blue', edgecolor ='white', linestyle ='dashed')
#     plt.title('Stocks')
#     plt.xlabel('Price')
#     plt.ylabel('Amount')
#     plt.show()
```
